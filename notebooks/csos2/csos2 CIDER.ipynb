```python
%matplotlib inline
from localcider.sequenceParameters import SequenceParameters
from localcider import plots
from Bio import SeqIO
from matplotlib import pyplot as plt
```

```python
fname = '/Users/flamholz/BDrive/Research/Savage/Csos2/covariation/csos2_hnea_IMG_homologs_12222016.fasta'
csos2s = []
for record in SeqIO.parse(fname, "fasta"):
    try:
        sp = SequenceParameters(record.seq.tostring())
        csos2s.append(sp)
    except:
        continue

fname = '/Users/flamholz/BDrive/Research/Savage/CarboxeColi/proteins/cbbL/RuBisCO_larges.fasta'
rbcLs = []
for record in SeqIO.parse(fname, "fasta"):
    try:
        sp = SequenceParameters(record.seq.tostring())
        rbcLs.append(sp)
    except:
        continue

fname = '/Users/flamholz/BDrive/Research/Savage/CarboxysomeEvolution/csosCA/csosCA_hnea_BLAST.txt'
csosCAs = []
for record in SeqIO.parse(fname, "fasta"):
    try:
        sp = SequenceParameters(record.seq.tostring())
        csosCAs.append(sp)
    except:
        continue

a_csos2 = csos2s[0]
a_rbcL = rbcLs[0]
a_csosCA = csosCAs[0]
all_sps = csos2s + rbcLs + csosCAs
```

```python
plots.show_multiple_phasePlot2(all_sps)
```

```python
plots.show_multiple_uverskyPlot2(all_sps)
```

```python
a_csos2.show_linearSigma(blobLen=20)
```

```python
a_csosCA.show_linearSigma(blobLen=20)
```

```python
a_rbcL.show_linearSigma(blobLen=20)
```

```python
kappas = [s.get_kappa() for s in all_sps]
lens = [len(s.get_sequence()) for s in all_sps]
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(lens, kappas)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('length (amino acids)', fontsize=16)
plt.ylabel('$\kappa$ parameter', fontsize=16)

plt.show()
```
